```python
def get_scam_type(text):
    text = str(text).lower()

    if 'otp' in text or 'verification code' in text:
        return 'OTP Scam'
    elif 'bank' in text or 'account' in text:
        return 'Banking Scam'
    elif 'lottery' in text or 'won' in text:
        return 'Internship Scam'
    elif 'job' in text or 'salary' in text:
        return 'Job Scam'
    elif 'link' in text or 'click' in text:
        return 'Phishing'
    else:
        return 'Lottery Scam'
```
